```python
import matplotlib.pyplot as plt

def T(n):
    count = 1
    while n != 1:
        if n % 2 == 0:
            n = n / 2
        else:
            n = (3 * n) + 1
        count += 1

    return count

X = range(1, 10000)
Y = []
for i in X:
    Y.append(T(i))

plt.scatter(X, Y, s=10)
plt.show
```
